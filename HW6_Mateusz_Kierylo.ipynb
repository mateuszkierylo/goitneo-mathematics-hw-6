**TASK 1**

You have ordered a study on employee work efficiency and received a functional dependency: the efficiency of employees when working on uniform tasks closely follows the formula y = -3*x**2 + 30*x

where y is the company's income generated by the developer, x is the number of working hours per day.

    1) Construct the function graph and the function's derivative graph on one plot.

    2) Calculate the maximum efficiency point and the efficiency value at this point analytically.

    3) Construct the graph of the work efficiency function and mark the point of maximum efficiency.

    4) Calculate the company benefit value that a developer brings during an 8-hour working day.Zrzut ekranu 2024-04-26 o 14.35.12.png

1)

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# define the variable and function:
x = sp.symbols('x')
f = -3*x**2 + 30*x

# calculate the derivative:
df = sp.diff(f, x)

print("The derivative of the function f(x) = -3*x**2 + 30*x with respect to x is:", df)
print('\n')


# define the function and its derivative:
def f(x):
    return -3*x**2 + 30*x

def df(x):
    return -6*x + 30

# denerate x values:
x = np.linspace(0, 10, 400)

# compute y values for function:
y_f = f(x)

# compute y derivative for function:
y_df = df(x)

# create the plot:
plt.figure(figsize=(8, 6))
plt.plot(x, y_f, label='y = -3x + 30x')
plt.plot(x, y_df, label='y = -6x + 30', linestyle=':', color='green')
plt.title('Function and its derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# maximum efficiency point calculated analytically:
x_max_efficiency = 5
y_max_efficiency = 75


# create the plot:
plt.figure(figsize=(8, 6))
plt.plot(x, y_f, label='y = -3x + 30x')
plt.scatter([x_max_efficiency], [y_max_efficiency], color='red')
plt.text(x_max_efficiency, y_max_efficiency - 7, f'Max\n({x_max_efficiency}, {y_max_efficiency})',
         horizontalalignment='center', color='red')
plt.title('Work efficiency function with maximum efficiency point')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# calculate the income for 8 hours of work using the efficiency function:

x_income = 8
y_income = f(x_income)

print("The income brought by an employee for 8 hours of work,\ncalculated using the function y_income is:", y_income)
```

**TASK 2**

You created a policy of interaction between employees, team leaders, and project managers and in result the function of employee work efficiency has changed. Now it looks like this: y = -3*(x-2)**2 + 30*x


Evaluate the efficiency after changing the work scheme for employees: how much more company income does one employee bring during the 8-hour working day?



1)

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# define the variable and function:
x = sp.symbols('x')
f = -3*(x-2)**2 + 30*x

# calculate the derivative:
df = sp.diff(f, x)

print("The derivative of the function f(x) = -3*(x-2)**2 + 30*x with respect to x is:", df)
print('\n')


# define the function and its derivative:
def f(x):
    return -3*(x-2)**2 + 30*x

def df(x):
    return 42 - 6*x

# denerate x values:
x = np.linspace(0, 10, 400)

# compute y values for function:
y_f = f(x)

# compute y derivative for function:
y_df = df(x)

# create the plot:
plt.figure(figsize=(8, 6))
plt.plot(x, y_f, label='y = -3(x-2) + 30x')
plt.plot(x, y_df, label='y = 42 - 6x', linestyle=':', color='green')
plt.title('Function and its derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```
